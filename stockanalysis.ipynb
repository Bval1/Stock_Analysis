Returns a DF with Cumulative Return for all Stocks

```python
def get_stock_df_from_csv(ticker):
    try:
        df = pd.read_csv(PATH + ticker + '.csv', index_col=0)
    except FileNotFoundError:
        print("File not found:", PATH + ticker + '.csv')
    else:
        return df

def get_cumulative_return_for_stocks(stock_df):
    tickers = []
    cumulative_returns = []
    for index, row in stock_df.iterrows():
        df = get_stock_df_from_csv(row['Ticker'])
        if df is None:
            pass
        else:
            tickers.append(row['Ticker'])
            cr = df['Cumulative Return'].iloc[-1]
            cumulative_returns.append(cr)
    return pd.DataFrame({'Ticker': tickers, 'Cumulative_Return': cumulative_returns})
```

```python
industrial = get_cumulative_return_for_stocks(industrial_df)
health_care = get_cumulative_return_for_stocks(health_df)
it = get_cumulative_return_for_stocks(it_df)
communication = get_cumulative_return_for_stocks(comm_df)
discretion = get_cumulative_return_for_stocks(discretion_df)
finance = get_cumulative_return_for_stocks(financial_df)
material = get_cumulative_return_for_stocks(material_df)
real_estate = get_cumulative_return_for_stocks(restate_df)
energy = get_cumulative_return_for_stocks(energy_df)
utility = get_cumulative_return_for_stocks(util_df)
```
